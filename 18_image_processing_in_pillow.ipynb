# Rotating images

```python
from PIL import Image 
img = Image.open("/content/drive/MyDrive/FINAL DATASET/new/5.png")

img_90_rot = img.rotate(90)
img_90_rot.save("/content/drive/MyDrive/FINAL DATASET/new/7.png")  #keeps original aspect ratio and dimensions

img_45_rot = img.rotate(45)
img_45_rot.save("/content/drive/MyDrive/FINAL DATASET/new/8.png")  #keeps original aspect ratio and dimensions

img_45_rot = img.rotate(45, expand=True)  #Dimensions are expanded to fir the entire image
img_45_rot.save("/content/drive/MyDrive/FINAL DATASET/new/9.png")  
```

```python
img_90_rot.show()
```

```python
img_45_rot.show()
```

#Flipping or transposing images

```python
from PIL import Image 
img = Image.open("/content/drive/MyDrive/FINAL DATASET/new/5.png")  #easy to see that the image is flipped

img_flipLR = img.transpose(Image.FLIP_LEFT_RIGHT)
img_flipLR.save("/content/drive/MyDrive/FINAL DATASET/new/9.png")

img_flipTB = img.transpose(Image.FLIP_TOP_BOTTOM)
img_flipTB.save("/content/drive/MyDrive/FINAL DATASET/new/10.png")
```

```python
img.show()
```

```python
img_flipLR.show()
```

```python
img_flipTB.show()
```

```python
img_flipLR.show()
```

# Color transforms, convert images between L (greyscale), RGB and CMYK

```python
# Color transforms, convert images between L (greyscale), RGB and CMYK
from PIL import Image 
img = Image.open("/content/drive/MyDrive/FINAL DATASET/new/5.png")

grey_img = img.convert('L')  #L is for grey scale
grey_img.save("/content/drive/MyDrive/FINAL DATASET/new/5.png")
```

```python
grey_img.show()
```
